```python
import camelot

def extract_table_to_csv(pdf_file, page_number, output_csv):
    try:
        # Extract table from the specified page
        tables = camelot.read_pdf(pdf_file, pages=str(page_number), flavor='stream')

        # Assuming the first table on the page is the one we want
        table_df = tables[0].df

        # Drop rows with any NaN values
        table_df = table_df.dropna()

        # Save the extracted data to CSV
        table_df.to_csv(output_csv, index=False)

        print(f"Data from page {page_number} extracted and saved to '{output_csv}'")

    except Exception as e:
        print(f"Error extracting table from page {page_number}: {e}")


pdf_file='/content/sunil_sbi.pdf'
output_csv = "page_34_table.csv"
extract_table_to_csv(pdf_file,34, output_csv)
```

```python
# Function to clean and preprocess the data
def clean_data(df):
    # Remove any commas and convert to numeric
    df = df.apply(lambda x: x.str.replace(',', '') if x.dtype == 'object' else x)
    df = df.apply(pd.to_numeric, errors='ignore')
    # Handle missing values
    df = df.fillna(0)
    return df

# Clean the data for Page 34
assets_page34 = clean_data(assets_page34)
liabilities_page34 = clean_data(liabilities_page34)

# Display the cleaned data
print("Cleaned Assets - Page 34:")
print(assets_page34)
print("\nCleaned Liabilities - Page 34:")
print(liabilities_page34)
```

```python
# Import the regular expression module
import re

def preprocess_text(text):
    # Remove parentheses, numbers, and special characters
    text = re.sub(r'[\(\)\d\W_]', ' ', text)

    # Convert to lowercase
    text = text.lower()

    return text

# Apply preprocessing to the 'Names' column
liabilities_page34['EqUITy AND LIABILITIES'] = liabilities_page34['EqUITy AND LIABILITIES'].apply(preprocess_text)
```

# MODEL BUILDING

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Drop rows with NaN values (if any)
liabilities_page34.dropna(inplace=True)


# Scale the features
scaler = StandardScaler()

X_liabilities_page34 = scaler.fit_transform(liabilities_page34[['2017']].values)
y_liabilities_page34 = liabilities_page34['2018'].values
```

```python
# Create the model for Assets
model_liabilities_page34 = LinearRegression()
model_liabilities_page34.fit(X_liabilities_page34, y_liabilities_page34)

# Predicting the next year's assets
next_year_liabilities_page34 = model_liabilities_page34.predict([X_liabilities_page34[-1]])

print("Predicted Liabilities for Next Year (Page 34):", next_year_liabilities_page34)
```

```python
# Drop rows with NaN values (if any)
assets_page34.dropna(inplace=True)
liabilities_page34.dropna(inplace=True)

# Scale the features
scaler = StandardScaler()

X_assets_page34 = scaler.fit_transform(assets_page34[['2017']].values)
y_assets_page34 = assets_page34['2018'].values
```
